```python
## So, what do qubits and quantum gate operations look like? Let's visualize this on a Bloch Sphere!
## A bloch sphere is a spherical representation of a qubit, where gate operations can be seen as rotations around each of the axes. 
```

```python
from qiskit.visualization import plot_bloch_vector

# We can plot the vectors using cartesian coordinates, where a 180 degree rotation corresponds to going from 0 to 1, or vice versa.
# Try changing each vector between -1 and 1 and see what you get!

plot_bloch_vector([0,1,0], title="New Bloch Sphere")
```

```python
import numpy as np
from qiskit.visualization import plot_bloch_vector
 
# You can use spherical coordinates instead of cartesian.
# These rotations correspond to angles phi, psi, and theta.
# Depending on how you build your circuit, what is actually trying to be minimized when you calculate expectation values is the energies
# corresponding to these angles (most often theta). In a variational algorithm like VQE, this is what you're trying to optimize.
# These angles correspond to wave amplitudes, which are related to the microwave pulses that are "injected" into the qubits.
# This is the underlying mechanism behind how quantum gates work. 
# Remember, in quantum mechanics, energy is measured through wave functions - Schrodinger's equation!
 
plot_bloch_vector([1, np.pi/2, np.pi/3], coord_type='spherical')
```

```python
## Let's amp things up a knotch and draw multi-qubit circuits on the bloch sphere.
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
 
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
 
state = Statevector(qc)
plot_bloch_multivector(state)
```
